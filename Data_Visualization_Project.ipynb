# Cancer Patient Dataset Analysis
## Dataset Description

The provided dataset, named cancer_patient_datasets.csv, contains information related to cancer patients. It consists of several columns representing different attributes or features of the patients. Here is a description of each column:

    1. Patient Id: A unique identifier for each patient.
    2. Age: The age of the patient in years.
    3. Gender: The gender of the patient (1 = male, 2 = female).
    4. Air Pollution Exposure: Level of exposure to air pollution (1 = none, 10 = very high)
    5. Alcohol Consumption: Level of alcohol intake (1 = none, 10 = very high)
    6. Dust Allergy Severity: Severity of dust allergy (1 = none, 10 = very severe)
    7. Occupational Hazard Exposure: Level of exposure to occupational hazards (1 = none, 10 = very high)
    8. Genetic Risk: Risk of developing cancer due to genetic factors (1 = low, 10 = very high)
    9. Chronic Lung Disease: Severity of chronic lung disease (1 = none, 10 = very severe)"
    10. Balanced Diet Adherence: Level of adherence to a balanced diet (1 = very poor, 10 = excellent)
    11. Obesity Level: Level of obesity (1 = healthy weight, 10 = very obese)
    12. Smoking Habits: Severity of smoking habits (1 = non-smoker, 10 = heavy smoker)
    13. Passive Smoking Exposure: Level of exposure to secondhand smoke (1 = none, 10 = very high)
    14. Chest Pain: Severity of chest pain (1 = none, 10 = very severe)
    15. Coughing Blood: Frequency of coughing blood (1 = never, 10 = very frequent)
    16. Fatigue Level: Level of fatigue (1 = none, 10 = extreme)
    17. Weight Loss Rate: Rate of weight loss (1 = none, 10 = very rapid)
    18. Shortness of Breath: Severity of shortness of breath (1 = none, 10 = very severe)
    19. Wheezing Frequency: Frequency of wheezing (1 = never, 10 = very frequent)
    20. Swallowing Difficulty: This factor can be addressed by adding a new category: "Severity of swallowing difficulty (1 = none, 10 = very severe)
    21. Clubbing of Fingernails: Severity of clubbing (1 = none, 10 = very severe)
    22. Frequency of Colds: Frequency of experiencing colds (1 = never, 10 = very frequent)
    23. Dry Cough: Severity of dry cough (1 = none, 10 = very severe)
    24. Snoring Frequency: Frequency of snoring (1 = never, 10 = very frequent)

    25. Level: The severity level of cancer (Low, Medium, High).

# Statistical Analysis
The One-Way Analysis of Variance (ANOVA) test

# Interpretation

    p-value < 0.05: Statistically significant difference in the means of the feature across different cancer severity levels.
    p-value >= 0.05: No statistically significant difference in the means of the feature across different cancer severity levels.

```python
from scipy.stats import f_oneway

features = ['OccuPational Hazards', 'Genetic Risk', 'Obesity', 'Smoking']

p_values = []

for feature in features:
    groups = [df[feature][df['Level'] == level] for level in df['Level'].unique()]
    anova_result = f_oneway(*groups)
    print(f"ANOVA result for {feature}: p-value = {anova_result.pvalue}")
    p_values.append(anova_result.pvalue)
```

 * #### All these p-values are extremely low (much lower than 0.05), indicating that the differences in means for these features across the different levels of cancer severity are highly significant. This means that each of these features has a significant effect on the cancer severity level.

# Feature Importance
show the past result as a graph

```python
log_p_values = [-np.log10(p) for p in p_values]# Transform the p-values using the negative logarithm (base 10)

# Visualization
plt.figure(figsize=(10, 6))
sns.barplot(y=features, x=log_p_values, orient='h')
plt.axvline(x=-np.log10(0.05), color='r', linestyle='--')  # Add a line to indicate the 0.05 threshold for significance
plt.title('ANOVA p-values for Different Features (Log Scale)')
plt.ylabel('Features')
plt.xlabel('-Log10(p-value)')
plt.show()
```

## Defining Bad Habits and Symptoms

```python
bad_habits_columns = ['Alcohol use', 'Smoking', 'Passive Smoker', 'Obesity', 'OccuPational Hazards']

bad_symptoms_columns = ['Chest Pain', 'Coughing of Blood', 'Fatigue', 'Weight Loss', 'Shortness of Breath', 
                        'Wheezing', 'Swallowing Difficulty', 'Clubbing of Finger Nails', 'Frequent Cold', 
                        'Dry Cough', 'Snoring']

df['Bad_Habits'] = df[bad_habits_columns].sum(axis=1)
df['Bad_Symptoms'] = df[bad_symptoms_columns].sum(axis=1)
```

## Create a donut chart for the Average Bad Symptoms by Cancer Severity Level

```python
bad_symptoms_avg = df.groupby('Level')['Bad_Symptoms'].mean()

plt.figure(figsize=(8, 8))
colors = sns.color_palette("husl", 3)
plt.pie(bad_symptoms_avg, labels=bad_symptoms_avg.index, colors=colors, autopct='%1.1f%%', startangle=90, wedgeprops={'width': 0.3}) # To create a donut chart
plt.title('Average Bad Symptoms by Cancer Severity Level')
plt.show()
```

# Feature Importance Analysis

Using machine learning models, you can analyze feature importance to understand which factors contribute most to predicting cancer severity.

```python
from sklearn.ensemble import RandomForestClassifier

X = df.drop(columns=['Patient Id', 'Level'])
y = df['Level']

X = pd.get_dummies(X, drop_first=True)# Encode categorical variables

model = RandomForestClassifier(n_estimators=100, random_state=42)# Train a Random Forest model
model.fit(X, y)

# Get feature importance
importances = model.feature_importances_
indices = np.argsort(importances)[::-1]

plt.figure(figsize=(15, 10))
plt.title('Feature Importance')
plt.bar(range(X.shape[1]), importances[indices], align='center')
plt.xticks(range(X.shape[1]), [X.columns[i] for i in indices], rotation=90)
plt.show()
```
